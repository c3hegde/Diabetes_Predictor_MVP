```python
# 1. Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import pickle

# 2. Load dataset
df = pd.read_csv('../Diabetes_Predictor_MVP/data/raw/diabetes.csv')

# 3. Preprocessing
X = df.drop('Outcome', axis=1)
y = df['Outcome']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 5. Train model
model = LogisticRegression()
model.fit(X_train, y_train)

# 6. Save model and scaler
pickle.dump(model, open('../Diabetes_Predictor_MVP/models/logistic_model.pkl', 'wb'))
pickle.dump(scaler, open('../Diabetes_Predictor_MVP/models/scaler.pkl', 'wb'))

# 7. Evaluate (Optional)
accuracy = model.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")
```
